# 1. Importing the required modules

```python
import numpy as np
import statistics
from IPython.display import clear_output
import math
import random
```

# 2. Part 1: Creating a stream

```python
sample_size = 20_000
stream_elements = random.sample(range(1, 1_000_000), sample_size)
```

```python
# Processing a stream of 1 billion would take hours
# Instead I'm using a smaller size
stream_size = 10_000_000
```

```python
def get_stream():
  for i in range(stream_size):
    idx = math.floor(random.uniform(1, sample_size))
    yield stream_elements[idx]
```

# 3. Part 2: Counting distinct numbers

```python
def trailing_zeros(n):
  bin_n = bin(n).replace("0b", "")
  s = str(bin_n)
  return len(s) - len(s.rstrip('0'))
```

```python
def flajolet_martin(stream_fn):
  max_zeros = -1
  i = 0
  for s in stream_fn():
    if i % 100_000 == 0:
      clear_output(wait=True)
      print(f'i: {i}, progress: {i*100/stream_size}%')
    max_zeros = max(max_zeros, trailing_zeros(s))
    i += 1

  return 2 ** max_zeros
```

```python
flajolet_martin(get_stream)
```

# 4. Part 3: AMS Algorithm

```python
def ams(stream_fn, num_start_times):
  start_times = random.sample(range(stream_size), num_start_times)
  item_counts = {}
  t = 0
  
  for i in stream_fn():
    if t % 100_000 == 0:
      clear_output(wait=True)
      print(f't: {t}, progress: {t*100/stream_size}%')
    if i in item_counts:
      item_counts[i] += 1
    elif t in start_times:
      item_counts[i] = 1
    
    t += 1
  
  return stream_size * statistics.mean([2 * c - 1 
                                        for c in item_counts.values()])
```

```python
ams(get_stream, 100)
```
