## 2. Are explicit songs more popular?

As we can see in the following cell, the mean of popularity for Explicit songs is higher so we can say they are more popular!

```python
df.groupby('Explicit').describe()['Popularity']
```

## 3. Sorting *Cermony* Songs

```python
df[df['Artist Name(s)'] == 'Ceremony'].sort_values('Track Number', ascending=False)
```

## 4. Top 10 frequent genres

```python
genres = df['Artist Genres'].apply(lambda x: str(x).split(","))
```

**Note:** `value_counts` return a sorted result

```python
genres.explode().value_counts()[:10]
```

## 5. Finding all songs which are related to rock

```python
artist_genres = df['Artist Genres'].fillna(False)
```

```python
mask = artist_genres.apply(lambda x: 'rock' in str(x).lower())
```

```python
df[mask]
```

## 8. Are Danceability and Energy of songs related?

Wikipedia says:

*In statistics, correlation or dependence is any statistical relationship, whether causal or not, between two random variables or bivariate data. In the broadest sense correlation is any statistical association, though it commonly refers to the degree to which a pair of variables are linearly related.*

We can use the `corr()` method to get the covariance matrix. As we can see there's a negative covariance between the two columns which means they are not correlated.

```python
df[['Danceability', 'Energy']].corr()
```
